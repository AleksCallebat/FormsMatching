```python
import urllib.request
import requests
%matplotlib inline
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
```

```python
# Get hippo
hippo_url = "https://i.ytimg.com/vi/96xC5JIkIpQ/maxresdefault.jpg"
urllib.request.urlretrieve(hippo_url, "test.jpg")

# Show image
plt.axis('off')
plt.imshow(Image.open('test.jpg'))
```

```python
url = "http://wincntkdemo.azurewebsites.net/api/uploader"
files = {'imagefile': open('test.jpg', 'rb')}
%time requests.post(url, files=files).json()
```
